```python

import pandas as pd
```

```python
def tweet_df(results):
    id_list= [tweet.id for tweet in results]
    dataset=pd.DataFrame(id_list, columns=['id'])
    
    dataset["text"]=[tweet.text for tweet in results]
    dataset['retweet']= [tweet.retweet for tweet in results]
    dataset['hashtag']=[tweet.entities.get('hashtag') for tweet in results]
    dataset['user_location']=[tweet.author.location for tweet in results]
    dataset['screen_name']=[tweet.author.screen_name for tweet in results]
    
    return dataset

dataset= tweet_df(results)
    
```

```python
text=dataset["text"]

for i in range(0, len(text)):
    txt=' '.join(word for word in text[i].split() if not word.startswith('https:'))
    dataset.set_value(i, 'text2', txt)

dataset.drop_duplicates('text2', inplace=True)
dataset.reset_index(drop= True, inplace=True)
dataset.drop('text', axis =1, inplace=True)
dataset.rename(columns={'text2': 'text'}, inplace=True)
```
